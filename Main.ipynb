```python
import numpy as np
import pandas as pd
from kmodes.kmodes import KModes
```

```python
data = np.random.choice(20, (100, 10))
```

```python
data
```

```python
dataset = pd.DataFrame({'a': data[:, 0], 'b': data[:, 1], 'c': data[:, 2], 'd': data[:, 3], 'e': data[:, 4], 'f': data[:, 5], 'g': data[:, 6], 'h': data[:, 7], 'i': data[:, 8], 'j': data[:, 9]})
```

```python
dataset
```

```python
km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1)
```

```python
clusters = km.fit_predict(data)
```

```python
kmodes = km.cluster_centroids_
shape = kmodes.shape
```

```python
print(kmodes)
```

```python
for i in range(shape[0]):
    if sum(kmodes[i,:]) == 0:
        print("\ncluster " + str(i) + ": ")
        print("no-skills cluster")
    else:
        print("\ncluster " + str(i) + ": ")
        cent = kmodes[i,:]
        for j in dataset.columns[np.nonzero(cent)]:
            print(j)
```
